```python
!python -m pip install paddlepaddle==2.4.2 -i https://pypi.tuna.tsinghua.edu.cn/simple

import paddle
from matplotlib import pyplot as plt
import numpy as np
import cv2

import os
from ppgan.apps import StylePredictor

%matplotlib inline

# load a art style image
style_image_path ='style.png'

# pretrained network weight path
weight_path ='style.pdparams'

# style name
style='starrynew'

# use cpu
cpu='cpu'
output_path='output_dir'
if cpu:
    paddle.set_device('cpu')
    
# initialize predict model
predictor = StylePredictor(output=output_path, style=style, weight_path=weight_path)

# test image datasets path
img_paths = os.listdir('test_imgs')
img_list=[]

# style transfer
for i in range(len(img_paths)):
    # test image path
    content_img_path = 'test_imgs/'+img_paths[i]
    print(content_img_path)
    
    # predicted
    _,stylized_visual = predictor.run(content_img_path, style_image_path)
    
    # result image save in list
    img_list.append(stylized_visual)
    
    # generated array image
    if (i+1)%25==0:
        vs1=np.hstack((img_list[0],img_list[1],img_list[2],img_list[3],img_list[4]))
        vs2=np.hstack((img_list[5],img_list[6],img_list[7],img_list[8],img_list[9]))
        vs3=np.hstack((img_list[10],img_list[11],img_list[12],img_list[13],img_list[14]))
        vs4=np.hstack((img_list[15],img_list[16],img_list[17],img_list[18],img_list[19]))
        vs5=np.hstack((img_list[20],img_list[21],img_list[22],img_list[23],img_list[24]))
        res=np.vstack((vs1,vs2,vs3,vs4,vs5))
        cv2.imwrite('res.jpg', res)
        res=cv2.imread('res.jpg', cv2.COLOR_BGR2RGB)

        print('display image...')
        plt.figure()
        plt.imshow(res)
        plt.show()
        img_list=[]
```
